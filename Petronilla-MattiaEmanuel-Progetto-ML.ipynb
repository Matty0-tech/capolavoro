In questo dataset abbiamo **51 valori** equivalenti ai 51 stati americani.
è diviso in **3 colonne** ognuna ad indicare rispettivamente:
- **Intensità dell'acqua**: quanta acqua viene usata per produrre energia.
- **Intensità delle emissioni**:  quante emissioni vengono prodotte per produrre energia
- **Consumo energetico totale (scalato).**

Inoltre dalla tabella sovrastante possiamo individuare **media**, **deviazione standard**, **valore massimo** e **minimo** e infine **mediana**


Già guardando questi dati senza l'aiuto di grafici possiamo notare la grande deviazione standard nellìintensità dell'acqua e ciò ci farà ipotizzare già la possibile presenza di valori **outlier** per la dispersione dei dati

```python
import pandas as pd
import matplotlib.pyplot as plt


df.columns = ['State', 'Water Intensity (m3/MWh)', 'Emission Intensity (m3/MWh)', 'Scaled Power Consumption (MWh)']



# Ordinamento per ciascun grafico
df_water = df.sort_values('Water Intensity (m3/MWh)', ascending=False)
df_emission = df.sort_values('Emission Intensity (m3/MWh)', ascending=False)
df_power = df.sort_values('Scaled Power Consumption (MWh)', ascending=False)

# Impostazioni grafiche
label_fontsize = 14
title_fontsize = 16
tick_fontsize = 12

# Creazione della figura
plt.figure(figsize=(14, 18)) 

# Grafico 1: Water Intensity
plt.subplot(3, 1, 1)
plt.bar(df_water['State'], df_water['Water Intensity (m3/MWh)'], color='blue')
plt.title("Water Intensity per Stato", fontsize=title_fontsize)
plt.xlabel("Stato", fontsize=label_fontsize)
plt.ylabel("m3/MWh", fontsize=label_fontsize)
plt.xticks(rotation=90, fontsize=tick_fontsize)
plt.yticks(fontsize=tick_fontsize)

# Grafico 2: Emission Intensity
plt.subplot(3, 1, 2)
plt.bar(df_emission['State'], df_emission['Emission Intensity (m3/MWh)'], color='brown')
plt.title("Emission Intensity per Stato", fontsize=title_fontsize)
plt.xlabel("Stato", fontsize=label_fontsize)
plt.ylabel("m3/MWh", fontsize=label_fontsize)
plt.xticks(rotation=90, fontsize=tick_fontsize)
plt.yticks(fontsize=tick_fontsize)

# Grafico 3: Power Consumption
plt.subplot(3, 1, 3)
plt.bar(df_power['State'], df_power['Scaled Power Consumption (MWh)'], color='lightgreen')
plt.title("Power Consumption per Stato", fontsize=title_fontsize)
plt.xlabel("Stato", fontsize=label_fontsize)
plt.ylabel("MWh", fontsize=label_fontsize)
plt.xticks(rotation=90, fontsize=tick_fontsize)
plt.yticks(fontsize=tick_fontsize)

plt.tight_layout()
plt.show()

```

Con questi **3 grafici a barre** possiamo andare ad approfondire i nostri dati vedendo le varie differenze di intensità di acqua, di emissione. E il consumo totale degli stati dell'America

con i **3 blox plot** possiamo andare ad analizzare graficamente:
- i valori di 1o e 3o quartile
- il valore della mediana
- con i baffi valori estremi con considerari outlier
- e con i pallini eventuali valori outlier se presenti

```python
import matplotlib.pyplot as plt

x = df['Water Intensity (m3/MWh)']
y = df['Emission Intensity (m3/MWh)']
sizes_raw = df['Scaled Power Consumption (MWh)']

sizes = sizes_raw / 10000

# Definiamo 3 range per le dimensioni
bins = [0, 1e6, 3e6, 1e8]  # esempio di soglie (puoi adattare)
colors_map = ['green', 'orange', 'red']

def get_color(value):
    if value < bins[1]:
        return 'green'
    elif value < bins[2]:
        return 'orange'
    else:
        return 'red'

colors = [get_color(v) for v in sizes_raw]

plt.figure(figsize=(12, 8))
plt.scatter(x, y, s=sizes, alpha=0.7, c=colors, edgecolors='k')

for i, state in enumerate(df['State']):
    plt.text(x.iloc[i], y.iloc[i], state, fontsize=9, ha='center', va='center', fontweight='bold')

plt.xlabel("Water Intensity (m3/MWh)", fontsize=14)
plt.ylabel("Emission Intensity (m3/MWh)", fontsize=14)
plt.title("Relazione tra Water Intensity e Emission Intensity", fontsize=16)

plt.grid(True)
plt.tight_layout()
plt.show()
```

Infine con questo **scatter plot (grafico a dispersione)** sono andato a confrontare tutte e 3 le colonne insieme così da vedere quanto uno stato con un consumo totale eccessivo **pallino rosso grande** consumasse.
Possiamo anche fare dei paragoni riguardante la colonna dell'acqua (asse delle x) e delle emissioni (asse delle y) comparato al consumo energetico totale.
E inoltre possiamo vefificare la dispersione notata in precedenza con la colonna dell'acqua notando dei pallini molto lontani dalla media.
